## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

## Task 1: Data Loading and Exploration

```python
df=pd.read_csv('box_office.csv')
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

## Task 2: Visualizing the Target Distribution

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sb.distplot(df['revenue'], hist=False, color='red')
plt.title('Revenue')
plt.subplot(1,2,2)
sb.distplot(np.log1p(df['revenue']), hist=False, color='blue')
plt.title('Log Transformed Revenue')
plt.show()
```

## Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16,8))
sb.jointplot(y=df['revenue'], x=df['budget'], kind='scatter', color='red')
sb.set(style='darkgrid')
plt.show()
```

```python
plt.figure(figsize=(16,8))
sb.jointplot(x=np.log1p(df['budget']), y=np.log1p(df['revenue']),kind='scatter', color='green')
sb.set(style='darkgrid')
```

## Task 4: Distribution of Languages in Film
